```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
live_data = pd.read_csv('live_data_week1.csv')
```

```python
live_data.head()
```

'''
loading or selecting features used for

'''

```python
cols_select = ['Total Time Spent on Website', 'Lead Origin', 'Lead Source',
       'Do Not Email', 'Last Activity', 'Specialization',
       'How did you hear about X Education', 'What is your current occupation',
       'Lead Quality', 'Lead Profile', 'City']

cate_cols = ['Lead Origin', 'Lead Source',
       'Do Not Email', 'Last Activity', 'Specialization',
       'How did you hear about X Education', 'What is your current occupation',
       'Lead Quality', 'Lead Profile', 'City']
```

```python
predictions_data = live_data[cols_select]
```

```python
predictions_data
```

```python
# Loading scaler model for Total Time Spent on Website	

with open('scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)

# Loading encoder model for categorical columns	

with open('label_encoders.pkl', 'rb') as f:
    label_encoders = pickle.load(f)


# loading ML model 

with open('ml_v1.pkl', 'rb') as f:
    ml = pickle.load(f)
```

```python
predictions_data['Total Time Spent on Website'] = scaler.transform(predictions_data[['Total Time Spent on Website']])
```

```python
for col in cate_cols:
    if col in label_encoders:
        predictions_data[col] = predictions_data[col].map(lambda s: label_encoders[col].transform([s])[0] 
                                              if s in label_encoders[col].classes_ 
                                              else -1)
```

```python
predictions_data.head()
```

```python
## getting ML predictions
predictions = ml.predict_proba(predictions_data)
```

```python
predictions = [prob[1] for prob in predictions]
```

```python
# combining with ID's
Prospect_ID = live_data['Prospect ID']
lead_status = predictions
share_data = pd.DataFrame({'user_id':Prospect_ID,'lead_status': lead_status})
```
